### Logarithmic scale

```python
fig, axes = plt.subplots(1, 2, figsize=(10,4))
x = np.linspace(0,5,11)     
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

### Ticks and Custom tick labels

```python
fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, x**2, x, x**3, lw=2)
ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=16)
yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=19) # use LaTeX formatted labels
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")
ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1, 1))
ax.yaxis.set_major_formatter(formatter)
```

### Axes, Numbers and label spacing

```python
# distance between x and y axis and the numbers on the axes
import matplotlib
matplotlib.rcParams['xtick.major.pad'] = 5
matplotlib.rcParams['ytick.major.pad'] = 5

fig, ax = plt.subplots(1, 1)
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("label and axis spacing")

# padding between axis label and axis numbers
ax.xaxis.labelpad = 5
ax.yaxis.labelpad = 5

ax.set_xlabel("x")
ax.set_ylabel("y")
```

```python
# restore defaults
matplotlib.rcParams['xtick.major.pad'] = 3
matplotlib.rcParams['ytick.major.pad'] = 3
```

### Axis position adjustments

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.subplots_adjust(left=0.15, right=.7, bottom=0.1, top=0.9);
```

#### Axis Grid

```python
fig, axes = plt.subplots(1, 2, figsize=(10,3))

# default grid appearance
axes[0].plot(x, x**2, x, x**3, lw=2)
axes[0].grid(True)

# custom grid appearance
axes[1].plot(x, x**2, x, x**3, lw=2)
axes[1].grid(color='brown', alpha=0.5, linestyle='--', linewidth=0.5)
```

### Axis spines

```python
fig, ax = plt.subplots(figsize=(6,2))

ax.spines['bottom'].set_color('blue')
ax.spines['top'].set_color('gray')

ax.spines['left'].set_color('red')
ax.spines['left'].set_linewidth(2)

# turn off axis spine to the right
ax.spines['right'].set_color("none")
ax.yaxis.tick_left() # only ticks on the left side
```

### Twin axes

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

#### Axes where x and y is zero(like Co-ordinate geometry)

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0)) # set position of x spine to x=0

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))   # set position of y spine to y=0

xx = np.linspace(-45, 30, 400)
ax.plot(xx, np.sin(xx));
```

### Other 2D Curves

```python
n = np.array([0,1,2,3,4,5])
xx = np.linspace(-0.75, 1., 100)
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))
axes[0].set_title("scatter")


axes[1].step(n, n**3, lw=2)
axes[1].set_title("step")

axes[2].bar(n, n**3, align="center", width=0.5, alpha=0.5)
axes[2].set_title("bar")

axes[3].fill_between(n, n**2, n**3, color="green", alpha=0.5);
axes[3].set_title("fill_between");
```

#### subplot2grid

```python
fig = plt.figure()
#axx = plt.subplot2grid(structure of number of subplots, position (row, column), colspan=number of columns in the structure that occupied by this subplot, rowspan=number of rows in the structure that occupied by this subplot)
ax1 = plt.subplot2grid((4,4), (0,0), colspan=3)
ax2 = plt.subplot2grid((4,4), (0,3), rowspan= 4)
ax3 = plt.subplot2grid((4,4), (1,1), rowspan= 2, colspan=2)
ax3.plot(np.arange(5), np.arange(5) **2)
ax3.text(2, 12, r"$square$", color="brown")
ax4 = plt.subplot2grid((4,4), (1,0))
ax5 = plt.subplot2grid((4,4), (2,0))
ax6 = plt.subplot2grid((4,4), (3,0), colspan= 2)
ax7 = plt.subplot2grid((4,4), (3,2))
fig.tight_layout()
```

#### gridspec

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure()
#gs = gridspec.GridSpec(rows, columns, height_ratios=ratios of height of each row , width_ratios= ratios of width of each row)
gs = gridspec.GridSpec(3, 2, height_ratios=[3,2,1], width_ratios=[1,2])

for g in gs:
    ax = fig.add_subplot(g)
fig.tight_layout()
```

#### add_axes

```python
fig, ax = plt.subplots()

ax.plot(xx, xx**2, xx, xx**3)
fig.tight_layout()

# inset
inset_ax = fig.add_axes([0.2, 0.56, 0.35, 0.35]) # X, Y, width, height

inset_ax.plot(xx, xx**2, xx, xx**3)
inset_ax.set_title('zoom near origin', fontdict={'fontsize':7})

# set axis range
inset_ax.set_xlim(-.2, .2)
inset_ax.set_ylim(-.005, .01)

# set axis tick locations
inset_ax.set_yticks([0, 0.005, 0.01])
inset_ax.set_xticks([-0.1,0,.1]);
```

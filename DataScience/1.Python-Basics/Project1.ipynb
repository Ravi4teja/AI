#                                      Student Grade System

```python
from statistics import mean

print("\nLogon")

UserDetails = {"Python": "1,Credential", "Admin": "Secret.1"}

x =  input("Username: ")

while(1):
    if x in UserDetails.keys():
        y = input("Password: ")
        if(y == UserDetails[x]):
            print("--- Login Successful ---")
            break
        else:
            print("***Invalid Password, will detonate in 5 seconds***")
            
    else:
        print("\n***Invalid UserName***\n")
        x = input("Username: ")    

print("""
Welcome to Grade Central

[1] - Enter Grades
[2] - Remove Student
[3] - Student Average Grades
[4] - Exit""")
print("What would you like to do today?")
Student_Grades = {}
def grade():
    while(1):
        try:
            student = input("Student name: ")
            if student not in Student_Grades.keys():
                Student_Grades[student] = []
            grade = int(input("Grade: "))
            print("Adding Grade...")
            Student_Grades[student].append(grade)
            print("Student Grades are: ", Student_Grades)
            break
        except:
            print("\n*** Please provide proper input: student name should contain Characters, Grade should be number ***")
    
def removeStudent():
    student = input("Student name: ")
    if student in Student_Grades.keys():
        print("Removing {}...".format(student))
        Student_Grades.pop(student)
        print("Student Grades are: ", Student_Grades)
    else:
        print(student, "not there in Grade central")

def studentAVGs():
    if len(Student_Grades.keys()) > 0:
        print("Average Grades for Students are", {i:mean(Student_Grades[i]) for i in Student_Grades.keys()})
    else:
        print("*** No Grades to calculate Student Average Grades ***")
        
while(1):
    try:
        N = int(input("\nEnter your choice: "))
        if( N>=1 and N<=4 ):
            if(N == 1):
                grade()
            if(N == 2):
                removeStudent()
            if(N == 3):
                studentAVGs()
            if(N == 4):
                print("Exiting...")
                break
        else:
            print("\n*** Please provide choice 1-4 ***")
    except:
        print("\n*** Please provide numeric choice 1-4 ***")
```

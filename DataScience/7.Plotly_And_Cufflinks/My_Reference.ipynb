# Plotly and Cufflinks

* Plotly is an interative visualization library.
* Cufflinks connects plotly with pandas.
* https://plot.ly/ is the website for plotly documentation and https://github.com/santosjorge/cufflinks for Cufflinks github repo 

```python
import pandas as pd
import numpy as np
import cufflinks as cf
%matplotlib inline
```

```python
from plotly import __version__
```

```python
print(__version__)
```

```python
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
#This will connect the JavaScript to notebook
init_notebook_mode(connected=True)
```

```python
#This is using cufflinks offline
cf.go_offline()
```

```python
df = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
```

```python
df.head()
```

```python
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [34, 42, 84]})
```

```python
df2
```

```python
df.plot()
```

```python
df.iplot()
#We can see the advantages of plotly
#saving as png, zooming, Pan, data on hover, controlling plots by selecting perticulat Column
```

### Scatter Plot

```python
df.iplot(kind='scatter', x='A', y='B')
#This is also a scatter plot with which all the markers are joined by lines
#we can use combination of ['lines', 'markers', 'text'] for mode property, by default it is 'lines'
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='markers')
```

```python
df.iplot(kind='scatter', x='A', y='B', mode='lines+markers', size=20)
```

### Bar Plot

```python
df2.iplot(kind='bar')
```

```python
df2.iplot(kind='bar', x='Category', y='Values')
```

```python
df.iplot(kind='bar')
```

```python
df.count().iplot(kind='bar')
```

```python
df.sum().iplot(kind='bar')
```

### Box Plots

```python
df.iplot(kind='box')
```

```python
df3 = pd.DataFrame({'x': [1, 2, 3, 4, 5], 'y': [10, 20, 30, 20, 10], 'z' : [5, 4, 3, 2, 1]})
```

```python
df3
```

### Surface Plot

```python
df3.iplot(kind='surface')
```

```python
df3.iplot(kind='surface', colorscale='rdylbu')
#check the plotly doc for colorscale codes
```

### Histograms

```python
df['A'].iplot(kind='hist', bins=30)
```

```python
df.iplot(kind='hist', bins=30)
```

### Spread Plot

```python
df[['A', 'B']].iplot(kind='spread')
#We can see the spread between plots
```

```python
df.iplot(kind='bubble', x='A', y='B', size='C')
#A Bubble plot is like a scatter plot except the size of the marker is based on the column given
```

```python
#Scatter Plot
pd.plotting.scatter_matrix(df, alpha=0.5)
```

# Pandas Built-in Data Visualization

Pandas has built-in capabilities for data visualization which built-off of matplotlib, but it baked into pandas for easier usage!

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('df1', index_col=0)
```

```python
df1.head()
```

```python
df2 = pd.read_csv('df2')
```

```python
df2.head()
```

```python
df1['A'].hist(bins=10)
```

```python
df1['A'].plot(kind='hist',bins=25)
```

```python
df1['A'].plot.hist(bins=20)
```

```python
df2.plot.area(alpha=0.25)
```

```python
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
df1.plot.line(y='B')
```

```python
    df1.plot.scatter(x='A', y='B')
```

```python
df1.plot.scatter(x='A', y='B', c='C')
#We can also use c parameter whose value can be displayed by the color scale
```

```python
df1.plot.scatter(x='A', y='B', c='C', cmap = 'coolwarm')
```

```python
df1.plot.scatter(x='A', y='B', s=df1['C']*30)
#We can also define 3rd column using s parameter whose value can be displayed using size of the point in the scatter plot
```

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns =['a', 'b'])
```

```python
df.plot.hexbin(x='a', y = 'b', gridsize=25, cmap='vlag')
#hexagonal binning plot
```

```python
df2['a'].plot.kde()
#Kernal density estimation plot
```

```python
df2['a'].plot.density()
```

```python
df2.plot.kde()
```

```python
df2.plot.box()
#Box plot
```
